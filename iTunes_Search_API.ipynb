```python
import requests
import json
```

```python
base_url = "https://itunes.apple.com/search"
```

```python
r = requests.get(base_url, params = {'term':'the beatles',\
                                     'country':'us',\
                                    'media':'music'})
print(json.dumps(r.json(), indent = 4))
```

```python
r.json().keys()
```

### Obtaining 200 results

```python
r = requests.get(base_url, params = {'term':'the beatles',\
                                     'country':'us',\
                                     'media':'music','limit':200})
print(json.dumps(r.json(), indent = 4))
```

```python
r.status_code
```

### Obtaining all Track Names and Release Dates

```python
info = r.json()
    
for i in range(len(info['results'])):
    print(info['results'][i]['trackName'] + " | "\
         + info['results'][i]['releaseDate'])
```

### Converting the results into a Data Frame

```python
import pandas as pd
```

```python
songs_df = pd.DataFrame(r.json()['results'])
songs_df
```

### Writing it to csv

```python
songs_df.to_csv("songs-df.csv")
```
